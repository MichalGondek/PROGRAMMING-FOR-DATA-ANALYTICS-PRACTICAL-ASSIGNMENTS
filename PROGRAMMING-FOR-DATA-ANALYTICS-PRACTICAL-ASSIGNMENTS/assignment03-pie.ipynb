```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download"
df = pd.read_csv(url)


domains = df['Email'].str.split('@').str[-1].value_counts()

domains = pd.concat([domains.head(8), pd.Series({'Other': domains[8:].sum()})])

plt.figure(figsize=(8,8))
plt.pie(domains, labels=domains.index, autopct='%1.1f%%', startangle=140,
        colors=plt.cm.Pastel1.colors, shadow=True)

plt.title('Email Domain Distribution', fontsize=15, fontweight='bold')
plt.show()
```
